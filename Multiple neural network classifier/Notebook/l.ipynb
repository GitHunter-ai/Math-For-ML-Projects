## Import libraries and predefined Modules

```python
# 
from utils.plot import plot_data, plot_decision_boundary
from model.nn import NeuralNetwork as nn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import numpy as np
import pandas as pd
np.random.seed(2)
```

## Load saved dataset (moons.csv)

```python
data=pd.read_csv(r"Data\moons.csv")
info=data.head(),data.info()
print(info)
```

## Reshaping data

```python
X=data[["x1","x2"]].to_numpy().reshape(-1,2)
Y=data.label.to_numpy().reshape(-1,1)
```

## Plot dataset

```python
# X: (1000, 2), Y: (1000, 1)

plot_data(X, Y)
```

## Train Model and plot boundary

```python
# 1. Train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=72)

# 2. Initialize and train the model
model = nn(n_h=10, learning_rate=1.8, num_iterations=5000)
model.fit(X_train, Y_train, print_cost=True)

# 3. Evaluate the model
acc = model.score(X_test, Y_test)
print(f"Test Accuracy: {acc:.2%} \n ")

# plot output
plot_decision_boundary(model.predict, X, Y)
```

## Second Dataset (circles.csv)

```python
data1=pd.read_csv(r"Data\circles.csv")

X1=data1[["x1","x2"]].to_numpy().reshape(-1,2)
Y1=data1.label.to_numpy().reshape(-1,1)
plot_data(X1, Y1)
```

```python

# # Train again
X_train1, X_test1, Y_train1, Y_test1 = train_test_split(X1, Y1, test_size=0.2, random_state=4)
model1 = nn(n_h=10, learning_rate=0.1, num_iterations=5000)
model1.fit(X_train1, Y_train1, print_cost=True)
print(f"Test Accuracy: {model.score(X_test1, Y_test1):.2%}")
plot_decision_boundary(model1.predict, X1, Y1)
```
